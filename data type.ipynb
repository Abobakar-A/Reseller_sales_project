## Data Wrangling 

```python
reseller_sales_wrangeld = (reseller_sales
                           .assign(Profit = lambda df_:df_['SalesAmount']-df_['TotalProductCost'],
                                  ProfitPrc = lambda df_:df_['Profit']/df_['SalesAmount'],
                                  SalesExtendedDiff= lambda df_:df_['SalesAmount'] - df_['ExtendedAmount'],
                                  OrderYear = lambda df_:df_['OrderDate'].dt.year,
                                   OrderMonth = lambda df_:df_['OrderDate'].dt.month,
                                   OrderDay= lambda df_:df_['OrderDate'].dt.day,
                                   OrderMonthName = lambda df_:df_['OrderDate'].dt.month_name().str[:3],
                                   OrderDayName = lambda df_:df_['OrderDate'].dt.day_name().str[:3],
                                   OrderQuarter = lambda df_:df_['OrderDate'].dt.quarter
                                  )
                                 
                           
                          )
reseller_sales_wrangeld
```

```python
# Repalce missing value in the product table 
products.info()
products_wrangeld =( products 
                   .assign(
                         ProductCategory  = lambda df_:df_['ProductCategory'].fillna('Missing'),
                         ProductSubcategory  = lambda df_:df_['ProductSubcategory'].fillna('Missing'),
                         status = lambda df_:df_['Status'].fillna('Missing')
                       )
                  )
products_wrangeld 
```

```python
# Wrangel the reseller data frame 
reseller_wrangeld1 = ( reseller 
                      .query('AnnualRevenue >=150000 and NumberEmployees >=35')
                      .query('CountryRegionName in ["Canada", "United States"]')
                      .assign(RevenuePerEmployee =lambda df_:df_['AnnualRevenue']/df_['NumberEmployees']
                          
                      )
    
)
reseller_wrangeld1
```

```python
# Use the inner join 
# The merge method 
reseller_products = (reseller_sales_wrangeld
                     .merge(products_wrangeld, how='inner', on = 'ProductKey')
                     .merge(reseller_wrangeld1, how='inner', on = 'ResellerKey')
                     
    
)
reseller_products.info()
```
